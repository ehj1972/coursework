First, we include the required modules:

```python
import requests
import pandas as pd
import numpy as np
import datetime
# Setting this option will print all columns of a dataframe
pd.set_option('display.max_columns', None)
# Setting this option will print all of the data in a feature
pd.set_option('display.max_colwidth', None)
```

Next, we send a GET request to the spacexdata website, and then pull the data. 
We then place it in a Pandas dataframe called data: 

```python
spacex_url="https://api.spacexdata.com/v4/launches/past"
response = requests.get(spacex_url)
data=pd.json_normalize(response.json())
```
